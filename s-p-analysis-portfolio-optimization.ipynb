##### Creating log returns Chart for the Top 10 using Yahoo Finance



*   Here is showing the movement of Price between the Top 10 from one point in time to another. It'll also be presented in numerical value



```python
!pip install yfinance
# Importing Yahoo Finance
import yfinance as yf

symbols = ['AAPL', 'GOOGL', 'GOOG', 'MSFT', 'AMZN', 'META', 'JPM', 'JNJ', 'XOM', 'BAC'] #FB will be change to META

# Create an empty DataFrame to store the log returns
log_returns = pd.DataFrame()

# Loop through the symbols and fetch data
for symbol in symbols:
    try:
        stock = yf.Ticker(symbol)
        historical_data = stock.history(period="1y")
        if not historical_data.empty:
            close_prices = historical_data['Close']
            log_returns[symbol] = np.log(close_prices / close_prices.shift(1))
        else:
            print(f"No data available for {symbol}")
    except Exception as e:
        print(f"Error fetching data for {symbol}: {str(e)}")

# Create subplots for log returns using Plotly Express
fig = px.line(log_returns, x=log_returns.index, y=log_returns.columns, labels={'index': 'Date', 'value': 'Log Returns'},
              title='Log Returns of Top 10 S&P 500 Companies')

# Show the subplots
fig.show()
```

##### Calculating the Annual Log Returns and adding it back into the Top 10 S&P500 Dataset

```python
# empty dictionary to store annual log returns
annual_log_returns = {}

# Looping through the symbols and fetch data for each company
for symbol in symbols:
    try:
        stock = yf.Ticker(symbol)
        historical_data = stock.history(period="1y")
        if not historical_data.empty:
            close_prices = historical_data['Close']
            log_returns = np.log(close_prices / close_prices.shift(1))

            # Calculate the annual log return and convert it to a percentage
            log_return_a = log_returns.mean() * 250 * 100

            # Store the result in the dictionary
            annual_log_returns[symbol] = log_return_a

            print(f"Annual Log Return for {symbol}: {log_return_a:.2f}%")
        else:
            print(f"No data available for {symbol}")
    except Exception as e:
        print(f"Error fetching data for {symbol}: {str(e)}")

```

#### Portfolio Optimization of the top 10

##### Installing the right essential libraries

```python
!pip install PyPortfolioOpt
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt import risk_models
from pypfopt import expected_returns
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices
from pypfopt import HRPOpt
from pypfopt.efficient_frontier import EfficientCVaR
from pypfopt.expected_returns import mean_historical_return
from pypfopt.risk_models import CovarianceShrinkage
```

##### Creating a for loop to gather historical Max Close $

```python
# Define a list of stock symbols
symbols = ['AAPL', 'GOOGL', 'GOOG', 'MSFT', 'AMZN', 'META', 'JPM', 'JNJ', 'XOM', 'BAC']

# Create an empty DataFrame to store the closing prices
closing_prices = pd.DataFrame()

# Loop through the symbols and fetch the previous close for each
for symbol in symbols:
    try:
        stock = yf.Ticker(symbol)
        historical_data = stock.history(period="Max")
        if not historical_data.empty:
            # Extract the 'Close' price from the historical data and store it in the DataFrame
            closing_prices[symbol] = historical_data['Close']
        else:
            print(f"No data available for {symbol}")
    except Exception as e:
        print(f"Error fetching data for {symbol}: {str(e)}")

# Assuming you have a DataFrame with a datetime index
# Let's call it 'closing_prices' for the sake of example

closing_prices.index = closing_prices.index.date

# Now, the index of the DataFrame will only contain the date portion
#print(closing_prices)
```

```python
closing_prices.tail()
```

##### Calculating the covariance matrix


*   The Sharpe ratio is the ratio between returns and risk. The lower the risk and the higher the returns, the higher the Sharpe ratio.
*   The higher the Sharpe ratio, the better the performance of the portfolio.



```python
u = expected_returns.mean_historical_return(closing_prices)
S = risk_models.sample_cov(closing_prices)

ef = EfficientFrontier(u, S)
weights = ef.max_sharpe() # optimisezes for maximal sharpe ratio
ef.portfolio_performance(verbose=True)
```

##### An investment amount of $,1000,000:



*   Here you can put your own specific number for investment for your portfolio
*   Here it's saying to invest 2300 shares in Google Class C and 42 shares in Exxon
*   Here we see that our Annual Expected Return is 21.7%



```python
# here using PyPortfolioOpt to tell us how much of each stock should be bought

latest_prices = get_latest_prices(closing_prices)

da = DiscreteAllocation(weights, latest_prices, total_portfolio_value=1000000)

allocation, leftover = da.greedy_portfolio() #making the most optimal decisions
print("Discrete allocation:", allocation)
print("Funds remaining: ${:.2f}".format(leftover))
#using the 1M$ it's suggested we buy the most in Google stocks etc
```

##### Hierarchical Risk Parity (HRP)



*   HRP as a technique for portfolio diversification where the assets are divided and weighted according to a hierarchical tree structure.
*   Expected Annual Return of 17.9% lower than the original 21.7%. However Sharpe Ratio has increased
*   HRP is not as sensitive to outliers
*   HRP suggests we put majority of our money into 218 shares into the **Bank of America** and as little as possible in **Big Tech**



```python
returns = closing_prices.pct_change().dropna()
hrp = HRPOpt(returns)
hrp_weights = hrp.optimize()
hrp.portfolio_performance(verbose=True)
print(dict(hrp_weights))
```

```python
da_hrp = DiscreteAllocation(hrp_weights, latest_prices, total_portfolio_value=100000)

allocation, leftover = da_hrp.greedy_portfolio()
print("Discrete allocation (HRP):", allocation)
print("Funds remaining (HRP): ${:.2f}".format(leftover))
```

##### Mean Conditional Value at Risk (mCVAR)

*   mCVAR works by measuring the worst-case scenarios for each asset in the portfolio
*   we should invest heavily in **META with 184 shares** and then **1 share in Bank of America**


```python
S = closing_prices.cov()
ef_cvar = EfficientCVaR(u, S)
cvar_weights = ef_cvar.min_cvar()

cleaned_weights = ef_cvar.clean_weights()
print(dict(cleaned_weights))
```

```python
da_cvar = DiscreteAllocation(cvar_weights, latest_prices, total_portfolio_value=100000)

allocation, leftover = da_cvar.greedy_portfolio()
print("Discrete allocation (CVAR):", allocation)
print("Funds remaining (CVAR): ${:.2f}".format(leftover))
```
