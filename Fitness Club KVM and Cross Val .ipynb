```python
# import the essential libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# import the df with some data processing 
df = pd.read_csv('databases/csvs/fitness_class_2212.csv')
column_to_drop = 'booking_id'
df = df.drop(column_to_drop, axis=1)
df = pd.get_dummies(df, columns=['category', 'time'])
df['weight']= df['weight'].fillna(df['weight'].mean())
df
```

```python
df['day_of_week'] = df['day_of_week'].str[:3]

day_mapping = {
    'Mon': 1,
    'Tue': 2,
    'Wed': 3,
    'Thu': 4,
    'Fri': 5,
    'Sat': 6,
    'Sun': 7
}

df['day_of_week'] = df['day_of_week'].map(day_mapping)
```

```python
df['days_before'] = df['days_before'].astype(str)
```

```python
df['days_before'] = df['days_before'].str.replace(' days', '').astype(int)
```

```python
df
```

```python
# X & y 
X = df.iloc[:, :4].values # : in python means a range neither has upper or lower bound. lower bound is 0
y = df.iloc[:, 4].values
```

```python
#train
from sklearn.model_selection import train_test_split, GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20, random_state= 42)
```

```python
# prevent overfitting 
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler() # remember this. the goal of standardization is to have all the values of the features in the same range. -3 & +3
X_train = sc.fit_transform(X_train) 
X_test = sc.transform(X_test) 
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel="rbf", random_state=0, degree =5)
classifier.fit(X_train, y_train)
```

```python
y_preds = classifier.predict(X_test)
print(np.concatenate((y_preds.reshape(len(y_preds),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
cm =confusion_matrix(y_test, y_preds)
print(cm)

acc=accuracy_score(y_test, y_preds)
print(acc) 
print(f" Model Accuracy: {acc * 100:.2f}%")
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score (estimator = classifier, X = X_train, y = y_train, cv =10) # cv = 10 (10 test folds)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100)) # remeber this code 
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
# here there is lee way between the accruacy of + or - 2.58
```
